```python
from sklearn.tree import DecisionTreeRegressor
import kagglehub
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
path = kagglehub.dataset_download("preslavrachev/bitcoin-historical-prices-from-cryptocompare-")
print("Path to dataset files:", path)
```

```python
# Menampilkan isi folder dataset untuk memastikan nama file yang benar
dataset_folder = '/root/.cache/kagglehub/datasets/preslavrachev/bitcoin-historical-prices-from-cryptocompare-/versions/2'



for filename in os.listdir(dataset_folder):
    print(filename)
```

```python
file_path = os.path.join(dataset_folder,'btc-prices-poloniex-23apr2017-05apr2018.csv')
```

```python
# Membaca file CSV yang ditemukan
df = pd.read_csv(file_path)

# Menampilkan 5 baris pertama dataset
print(df.head())
```

```python
# Menampilkan nama-nama kolom dalam dataset
print(df.columns)
```

```python
x = df[['timestamp', 'open', 'close', 'low', 'high', 'weightedAverage', 'volume', 'quoteVolume']]
y = df['close']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state = 42)
```

```python
dtree = DecisionTreeRegressor(max_depth=3)
dtree.fit(x_train, y_train)
```

```python
accuracy = dtree.score(x_test, y_test)
print(f"Akurasi model: {accuracy*100:.2f}%")
```

```python
plt.figure(figsize=(30, 20))
plot_tree(dtree, filled=True)
plt.show()
```
